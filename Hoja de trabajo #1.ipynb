### Problema #3: Imagen a color a 3 en escala de grises

```python
def imagenes_gris (imagen):
    alto = imagen.shape[0]
    ancho = imagen.shape[1]
    gris_red = np.zeros((alto,ancho,1))
    gris_green = np.zeros((alto,ancho,1))
    gris_blue = np.zeros((alto,ancho,1))
    for i in range(0, alto):
        for j in range (0, ancho):
            pixel = imagen[i,j]

            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]

            gris_red[i,j] = int(0*blue + 0*green + 1*red)
            gris_green[i,j] = int(0*blue + 1*green + 0*red)
            gris_blue[i,j] = int(1*blue + 0*green + 0*red)
            
            

    cv2.imwrite("gris_red.jpg",gris_red)
    cv2.imwrite("gris_green.jpg",gris_green)
    cv2.imwrite("gris_blue.jpg",gris_blue)
    print('Listo. Las imágenes fueron exportadas a la carpeta')
```

```python
#Prueba de la función
imagen3 = cv2.imread("Oslo.jpg")
imagenes_gris (imagen3)
```

### Problema #4: Histogramas

```python
def histogramas (imagen):
    imagen = skimage.io.imread(fname="Oslo.jpg")
    fig, ax = plt.subplots()
    plt.imshow(imagen)
    plt.show()

    # Tupla para seleccinar colores de cada canal
    colors = ("red", "green", "blue")
    channel_ids = (0, 1, 2)

    # Creamos el histograma con una línea para cada color
    plt.figure()
    plt.xlim([0, 256])
    for channel_id, c in zip(channel_ids, colors):
        histogram, bin_edges = np.histogram(imagen[:, :, channel_id], bins=256, range=(0, 256))
        plt.plot(bin_edges[0:-1], histogram, color=c)

    plt.title("Histograma de RGB")
    plt.xlabel("Valor de color")
    plt.ylabel("Conteo de pixeles")

    plt.show()
```

```python
#Prueba de la función
imagen4 = cv2.imread("Oslo.jpg")
histogramas(imagen4)
```

### Problema #5: Escala de grises ponderada

### Escala de grises ponderada

Dado que el ojo humano es más sensible al color verde y menos sensible al azul, los valores de los colores se ponderan para que el resultado obtenido sea más estético visualmente.
Con la aproximación del promedio ponderado, al transformar una imagen a color hacia una escala de grises cada valor de la matriz RGB se multiplica por un índice establecido (peso) y se obtiene así una matriz resultado.
Una propuesta podrían ser las siguientes ponderaciones:

**<center> (0.299 x blue) + (0.587 x green) + (0.114 x red) </center>**
Según lo visto en clase
    


```python
def gris_ponderado (imagen):
    alto = imagen.shape[0]
    ancho = imagen.shape[1]
    gris_ponderado = np.zeros((alto,ancho,1))
    for i in range(0, alto):
        for j in range (0, ancho):
            pixel = imagen[i,j]

            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]

            gris_ponderado[i,j] = int(0.11*blue + 0.59*green + 0.3*red)
            #aquí se usó una ponderación recomendada en una referencia de internet

    cv2.imwrite("gris_ponderado.jpg",gris_ponderado)
    gris_ponderado = cv2.imread("gris_ponderado.jpg")
    gris_ponderado = cv2.cvtColor(gris_ponderado,cv2.COLOR_BGR2RGB)
    plt.imshow(gris_ponderado)
    plt.show()
```

```python
#Prueba de la función
imagen = cv2.imread("Oslo.jpg")

gris_ponderado(imagen)
```

### Problema #6: HSV

### Espacio de color HSV

El espacio HSV (por sus siglas en inglés  *Hue, Saturation, Value*) define los colores en función de tres factores: Matiz, saturación y valor.

 - **Matiz (H):** Se representa como un grado de ángulo cuyos valores posibles van de 0 a 360° (aunqueen algunas ocasiones se normalizan del 0 al 100%). Cada valor corresponde a un color específico. Ejemplos: 0 es rojo, 60 es amarillo y 120 es verde.

 - **Saturación (S):** mide la pureza de un color, es decir la ausencia de otros colores. Dado entre 0 y 1, 0 representa sin saturación alguna (blanco), hasta 1 que sería el matiz en toda su intensidad.

 - **Valor o Brillo (V):** hace referencia a la intensidad de la luz, es decir, la cantidad de luz que hay en ese color, desde un 0% (negro) al 100%.
